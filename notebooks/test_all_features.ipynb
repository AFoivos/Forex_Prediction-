```python
import os
import sys

project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)
    
from data_handle import(
    ForexDataClean,
    ForexDataLoad
)

from features import ForexFeauturesExtractor

from plots import ForexPlotter

import warnings
warnings.filterwarnings('ignore')

data = ForexDataLoad(file_path = '../data/xauaud.csv').data

cleaner = ForexDataClean(data = data)
data = cleaner.fast_cleaner()
```

```python
features = ForexFeauturesExtractor(
    data = data
)
```

```python
indicators_data, signals_data, label_data = features.extract_all_features()
```

```python
for signal, df in signals_data.items():
    print('=' * 70)
    print(signal, df.shape,df.isna().sum().sum())
```

```python

for key in signals_data.keys():
    print('=' * 70)
    print(signals_data[key].columns,'\n index name =', signals_data[key].index.name)
```

```python
data = {
    'indicators_data' : indicators_data,
    'signals_data' : signals_data,
    'label_data' : label_data
}

min_date = min(indicators_data.index)
max_date = max(indicators_data.index)

for name, df in data.items():
    if name != 'signals_data':
        if max_date != max(df.index) or min_date != min(df.index):
            print('!!!!!!!!!!!!!!!!!!!!!!')
        print(f'shape of {name}: {df.shape}, min date: {min(df.index)}, max date {max(df.index)} ')
    else:
        for name, df in signals_data.items():
            if max_date != max(df.index) or min_date != min(df.index):
                print('!!!!!!!!!!!!!!!!!!!!!!!!!!')
            print(f'shape of {name}: {df.shape}, min date: {min(df.index)}, max date {max(df.index)} ')
```
